# Python codes for finding relative abundance

```python


def Relative_abundance(data):
    G=nx.read_weighted_edgelist(data)
    A=nx.to_numpy_array(G)
    F = np.zeros(15)
    n = A.shape[0]  # nodes
    k = np.dot(A, np.ones(n))  # degrees
    A2 = np.dot(A, A)
    A3 = np.dot(A2, A)
    # A4 = np.dot(A3, A)
    # A5 = np.dot(A4, A)
    A2b = A2 - diags(k.flatten(), 0, (n, n))
    F10s = np.sum(A2b * A2b - A2b, axis=1) / 2
    t = np.diag(A3) / 2
    tk = t * (k - 2)
    kk = k * (k - 1)
    m = np.sum(k) / 2
    A2A = A2 * A

    F[1] = np.sum(kk) / 2  # |P2|
    F[2] = np.sum(t) / 3  # |C3|
    F[3] = np.sum((k - 1) * (np.dot(A, (k - 1)))) / 2 - F[1] * 3  # |P3|
    F[4] = np.sum(kk * (k - 2)) / 6  # |S13|
    # F[4] = (np.sum(np.diag(A4)) - 4 * F[0] - 2 * m) / 8  # |C4|
    F[5] = np.sum(tk)  # |T31|
    F[7] = np.sum(A2A * A2A - A2A) / 4  # |Diamond|
    # F[8] = np.sum(np.diag(A5)) / 10 - 3 * F[1] - F[6]  # |C5|
    F[9] = np.sum(tk * (k - 3)) / 2  # |Cr|
    F[10] = np.sum((k - 2) * F10s.reshape(-1, 1)) - 2 * F[7]  # |T41|
    F[11] = np.sum((k - 2) * (A2 * A) * (k - 2)) / 2 - 2 * F[7]  # |Reindeer|
    F[12] = np.sum(t * (np.sum(A2, axis=0) - k)) - 6 * F[1] - 2 * F[6] - 4 * F[7]  # |T32|
    F[13] = np.dot(t, t - 1) / 2 - 2 * F[7]  # |Bowtie|
    F[14] = np.sum(np.sum(A * A2 * A3)) / 2 - 9 * F[1] - 2 * F[6] - 4 * F[7]  # |House|
    
    
    Count =np.array([F[0],F[1],F[2],F[3],F[6],F[7],F[9],F[10],F[11],F[12],F[13],F[14]])
#     return Coun
    Trials = 50 # Average over at least 50 random graphs
    for i in range(Trials):
        R=nx.erdos_renyi_graph(n,2*m/n/(n-1))  # Here we create a random graph
        AR=nx.to_numpy_array(R)    # The rest of the loop just makes use of the formulae calculate earlier
        # AR=np.array(AR)
        kR = np.dot(AR, np.ones(n))  # degrees
        #mR = np.sum(kR)/2
        FR = np.zeros(15)
        #n = A.shape[0]  # nodes
        kR = np.sum(AR, axis=1)  # degrees
        A2R = AR @ AR
        A3R = A2R @ AR
        # A4 = A3 @ A
        # A5 = A4 @ A
        A2bR = A2R - diags(kR.flatten(), 0, (n, n))
        F10Rs = np.sum(A2bR * A2bR - A2bR, axis=1) / 2
        tR = np.diag(A3R) / 2
        tkR = tR * (kR - 2)
        kkR = kR * (kR - 1)
        mR = np.sum(kR) / 2
        A2AR = A2R * AR

        FR[0] = np.sum(kkR) / 2  # |P2|
        FR[1] = np.sum(tR) / 3  # |C3|
        FR[2] = np.sum((kR - 1) * (np.dot(AR, (kR - 1)))) / 2 - FR[1] * 3  # |P3|
        FR[3] = np.sum(kkR * (kR - 2)) / 6  # |S13|
        # F[4] = (np.sum(np.diag(A4)) - 4 * F[0] - 2 * m) / 8  # |C4|
        FR[6] = np.sum(tkR)  # |T31|
        FR[7] = np.sum(A2AR * A2AR - A2AR) / 4  # |Diamond|
        # F[8] = np.sum(np.diag(A5)) / 10 - 3 * F[1] - F[6]  # |C5|
        FR[9] = np.sum(tkR * (kR - 3)) / 2  # |Cr|
        FR[10] = np.sum((kR - 2) * F10Rs.reshape(-1, 1)) - 2 * FR[7]  # |T41|
        FR[11] = np.sum((kR - 2) * (A2R * AR) * (kR - 2)) / 2 - 2 * FR[7]  # |Reindeer|
        FR[12] = np.sum(tR * (np.sum(A2R, axis=0) - kR)) - 6 * FR[1] - 2 * FR[6] - 4 * FR[7]  # |T32|
        FR[13] = np.dot(tR, tR - 1) / 2 - 2 * FR[7]  # |Bowtie|
        FR[14] = np.sum(np.sum(AR * A2R * A3R)) / 2 - 9 * FR[1] - 2 * FR[6] - 4 * FR[7]  # |House|


    CountR =np.array([FR[0],FR[1],FR[2],FR[3],FR[6],FR[7],FR[9],FR[10],FR[11],FR[12],FR[13],FR[14]])/Trials
            #abund_RE=(Count-CountR)/(Count+CountR)
    RE_ABUN=(Count-CountR)/(Count+CountR)
    return RE_ABUN
```

# Python codes for finding bipartivity measures for random graphs

```python


def network(net):
    if net[-3:] == 'csv':
        import csv
        with open(net[0:-3]+'txt', "w") as my_output_file:
            with open(net, "r") as my_input_file:
                [my_output_file.write(" ".join(row) + '\n') for row in csv.reader(my_input_file)]
            my_output_file.close()
    elif net[-3:] == 'txt':
        G = nx.read_weighted_edgelist(net)  # Load in a file
        A = nx.to_numpy_array(G)
        # Eigenvalue and Eigenvector of A

        bλ_r = []
        bs_r = []
        be_r = []

        Trials = 100  # Average over at least 100 random graphs
        for i in range(Trials):
            n = len(G.nodes)
            m = len(G.edges)
            #e = np.ones((m, 1))
            
            R = nx.erdos_renyi_graph(n, 2*m/n*(n-1))  # Here we create a random graph
            AR = nx.to_numpy_array(R)  # The rest of the loop just makes use of the formulae calculate earlier
            AR = AR
            #kR = AR @ e

            LambdaR, QR = lg.eigh(AR)

            # maximum and minimum eigenvalues of A.
            maximum_LambdaR = np.max(LambdaR)
            minimum_LambdaR = np.min(LambdaR)
            # spectral measure of bipartivity

            bλR = -minimum_LambdaR / maximum_LambdaR
            bλ_r.append(bλR)
            bl = np.mean(bλ_r)

            cosh_LambdaR = np.sum(np.cosh(LambdaR))  # sum of cosh(λ)
            exp_LambdaR = np.sum(np.exp(LambdaR))  # sum of exp(λ)

            bsR = cosh_LambdaR / exp_LambdaR
            bs_r.append(bsR)
            bs = np.mean(bs_r)

            beR = np.sum(np.exp(-LambdaR)) / np.sum(np.exp(LambdaR))
            be_r.append(beR)
            be = np.mean(be_r)

        return [bl,bs,be]
 

   
```
